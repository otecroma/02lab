```python
import numpy as np   

import matplotlib.pyplot as plt

import time      #імпортуємо time для роботи з часом

import tracemalloc   #інструмент для відстеження і аналізу використання пам'яті

random_10_int_array = np.random.randint(0, 100, size=10) 
random_10_float_array = np.random.uniform(0.1, 99.9, size=10) 
 
random_100_int_array = np.random.randint(0, 100, size=100) 
random_100_float_array = np.random.uniform(0.1, 99.9, size=100) 
 
random_1000_int_array = np.random.randint(0, 100, size=1000) 
random_1000_float_array = np.random.uniform(0.1, 99.9, size=1000) 

 
random_10000_int_array = np.random.randint(0, 100, size=10000) 
random_10000_float_array = np.random.uniform(0.1, 99.9, size=10000) 
 

arrays = [ 
    ("10 Integers", random_10_int_array), 
    ("10 Floats", random_10_float_array), 
    ("100 Integers", random_100_int_array), 
    ("100 Floats", random_100_float_array), 
    ("1000 Integers", random_1000_int_array), 
    ("1000 Floats", random_1000_float_array), 
    ("10000 Integers", random_10000_int_array), 
    ("10000 Floats", random_10000_float_array) 
] 
```

```python
def reverse_array(array):
    # Перевірка, чи масив має лише один елемент; якщо так, повернути сам масив
    if len(array) == 1:
        return array
    
    # Якщо в масиві більше одного елементу, створити пустий список для зберігання реверсованих елементів
    reversed_array = []
    
    # Проходження по масиву в зворотньому порядку за допомогою циклу for
    # Починаємо з останнього елемента (len(array) - 1) і зменшуємо індекс на 1 на кожній ітерації до досягнення 0
    for i in range(len(array) - 1, -1, -1):
        # Додаємо кожний елемент з оригінального масиву до реверсованого масиву
        reversed_array.append(array[i])
    
    # Повернення реверсованого масиву
    return reversed_array
```

```python
def bubble_sort(l):
    # Проходиться по списку l, починаючи з кінця і закінчуючи після першого елемента
    # Для кожного проходу по списку, ми переглядаємо список і обмінюємо пари сусідніх елементів, якщо вони не впорядковані
    for passesLeft in range(len(l)-1, 0, -1):
        # Внутрішній цикл, що проходиться по елементам з індексами від 0 до passesLeft
        for index in range(passesLeft):
            # Порівнюємо поточний елемент з наступним
            if l[index] > l[index + 1]:
               # Якщо поточний елемент більший за наступний, міняємо їх місцями
               l[index], l[index + 1] = l[index + 1], l[index]
    # Повертаємо відсортований список l
    return l
```

```python
def insertion_sort(array):
    # Проходимо по всіх елементах масиву, починаючи з другого (індекс 1)
    for i in range(1, len(array)):
        # Зберігаємо поточний елемент масиву
        current = array[i]
        # Встановлюємо індекс j на один елемент лівіше від поточного елемента
        j = i - 1
        # Поки j не стане менше нуля та поточний елемент менше елемента на позиції j,
        # ми зсуваємо всі елементи, які більші за поточний, на одну позицію вправо
        while j >= 0 and current < array[j]:
            array[j + 1] = array[j]
            j -= 1
        # Вставляємо поточний елемент на його правильне місце
        array[j + 1] = current
    # Повертаємо відсортований масив
    return array
```

```python
def quick_sort(arr):
    # Базовий випадок: якщо масив містить один елемент або менше, він вже відсортований
    if len(arr) <= 1:
        return arr

    # Вибираємо елемент опори, це може бути будь-який елемент, але тут обирається середній
    pivot = arr[len(arr) // 2]

    # Створюємо три підмасиви для елементів менших, рівних та більших за опорний елемент
    low = []
    high = []
    equal = []

    # Розподіляємо елементи в масиві на ці три підмасиви
    for num in arr:
        if num < pivot:
            low.append(num)
        elif num > pivot:
            high.append(num)
        else:
            equal.append(num)

    # Рекурсивно сортуємо підмасиви менших та більших елементів
    # Потім з'єднуємо їх у порядку low, equal, high
    return quick_sort(low) + equal + quick_sort(high)
```

```python
def binary_search(array, item):
    # Ініціалізуємо змінні для нижньої та верхньої меж пошуку
    low = 0
    high = len(array) - 1

    # Поки нижня межа не перевищує верхню межу
    while low <= high:
        # Обчислюємо середину діапазону
        mid = (low + high) // 2
        # Значення, яке ми знаходимо в середині діапазону
        guess = array[mid]
        # Якщо значення співпадає зі шуканим елементом, повертаємо його індекс
        if guess == item:
            return mid
        # Якщо значення менше за шуканий елемент, зсуваємо нижню межу вправо
        elif guess < item:
            low = mid + 1
        # Якщо значення більше за шуканий елемент, зсуваємо верхню межу вліво
        else:
            high = mid - 1
    # Якщо елемент не знайдено, повертаємо -1
    return -1
```

```python
import time
import tracemalloc

def measure_algorithm(algorithm, array, *args):
    # Початок вимірювання часу
    start_time = time.time()
    # Початок відстеження виділення пам'яті
    tracemalloc.start()
    
    # Виконуємо алгоритм з переданими аргументами
    if algorithm == binary_search:
        # Для бінарного пошуку аргументи передаються як *args
        result = algorithm(array, *args)
    else:
        # Для інших алгоритмів не потрібні додаткові аргументи
        result = algorithm(array)
    
    # Зупиняємо вимірювання часу
    end_time = time.time()
    # Розраховуємо час виконання
    time_of_execution = end_time - start_time
    
    # Отримуємо поточне та пікове використання пам'яті
    current, peak = tracemalloc.get_traced_memory()
    # Зупиняємо відстеження виділення пам'яті
    tracemalloc.stop() 
    
    # Повертаємо час виконання та пікове використання пам'яті
    return time_of_execution, peak
```

```python

import matplotlib.pyplot as plt

# Створення списку з алгоритмами, які ми хочемо порівняти
algorithms = [reverse_array, bubble_sort, insertion_sort, quick_sort, binary_search]

# Створення фігури та осей для графіків
fig, axs = plt.subplots(5, figsize=(10, 20))

# Проходимося по кожному алгоритму
for j, algorithm in enumerate(algorithms):
    time_taken_list = []
    memory_used_list = []
    # Проходимося по кожному масиву
    for array_name, array in arrays:
        # Вимірюємо час виконання та використання пам'яті для поточного алгоритму та масиву
        if algorithm == binary_search:
            time_taken, memory_used = measure_algorithm(algorithm, quick_sort(array), 14)
        else:
            time_taken, memory_used = measure_algorithm(algorithm, array)
        # Виводимо вимірювані дані для кожного алгоритму та масиву
        print(f"{algorithm.__name__} with {array_name}: Time taken - {time_taken} seconds, Memory used - {memory_used} ")
        # Додаємо час виконання та використання пам'яті до відповідних списків
        time_taken_list.append(time_taken)
        memory_used_list.append(memory_used)

    # Побудова графіку для часу виконання
    axs[j].plot([name for name, _ in arrays], time_taken_list, label='Time Taken (s)', marker='o')
    axs[j].set_xlabel('Array Size and Type')
    axs[j].set_ylabel('Time (s)')
    axs[j].set_title(f'{algorithm.__name__} Performance')
    axs[j].legend()

    # Додаткова ось для графіку використання пам'яті
    axs_mem = axs[j].twinx()
    axs_mem.plot([name for name, _ in arrays], memory_used_list, color='orange', label='Memory Used ', marker='o')
    axs_mem.set_ylabel('Memory ')
    axs_mem.legend()

# Виведення графіків
plt.tight_layout()
plt.show()
```
